##password_strenght


```python
import re

def password_strength(password):
    if len(password) < 8:
        return "Débil: La contraseña debe tener al menos 8 caracteres."
    if not re.search(r"\d", password):
        return "Débil: La contraseña debe contener números."
    if not re.search(r"[A-Z]", password):
        return "Débil: La contraseña debe tener al menos una letra mayúscula."
    if not re.search(r"[!@#$%^&*()_+]", password):
        return "Débil: La contraseña debe contener al menos un carácter especial."
    return "Fuerte: La contraseña es segura."

# Ejemplo de uso
password = "Contraseña123!"
print(password_strength(password))
```
